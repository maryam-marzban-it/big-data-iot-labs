```python
# vi läser in csv filen
df_home = pd.read_csv("raw_energydata.csv")

df_home.size

# kollar på första raderna
df_home.head()
```

```python
# datatyper,minne och nullvärden
df_home.info()

# statistik sammanfattning
df_home.describe()


# antal rader och kolumner
df_home.shape

# kolumnnamn
df_home.columns
```

```python
# koden printar en sample på 3 slumpmässiga datapunkter
print("\nExempelrader:\n", df_home.sample(3))
```

```python
# tittar på datakvalitet
# Problem 1: Saknade värden
saknade_per_kolumn = df_home.isnull().sum()
saknade_totalt = saknade_per_kolumn.sum()
print(f" SAKNADE VÄRDEN: {saknade_totalt} totalt")
for kolumn, antal in saknade_per_kolumn[saknade_per_kolumn > 0].items():
    print(f"  • {kolumn}: {antal} saknade")
```

```python
# Problem 2: Dubbletter
dubbletter = df_home.duplicated().sum()
print(f"\n DUBBLETTER: {dubbletter} identiska rader")
```

```python
# Problem 3: Outliers (extrema värden)
print(f"\n EXTREMA VÄRDEN:")
# Kök-temperatur
temp_kök = df_home['Temp_Kök_C'].dropna()
temp_för_höga = (temp_kök > 50).sum()
print(f"  • Kök-temp över 50°C: {temp_för_höga} (omöjligt!)")
```

```python
# Energi
energi = df_home['Vitvaror_Energi_Wh'].dropna()
energi_negativ = (energi < 0).sum()
print(f"  • Negativ energi: {energi_negativ} (fysiskt omöjligt!)")
```

```python
# Fuktighet
fukt = df_home['Fukt_Badrum_Procent'].dropna()
fukt_över_100 = (fukt > 100).sum()
print(f"  • Fuktighet över 100%: {fukt_över_100} (omöjligt!)")
```

```python
#visualisering av energyförbrukning över tid
df_home.head(1000)['Vitvaror_Energi_Wh'].plot(kind='line',title='Energiförbrukning över tid')
```

```python
print(f"\nTOTALT: Data har {saknade_totalt + dubbletter + temp_för_höga + energi_negativ + fukt_över_100} problem!")
```

```python
# här printas först medelvärden på energyförbrukning
print(df_home['Vitvaror_Energi_Wh'].mean().round(2))

# min och max värden
print(df_home['Vitvaror_Energi_Wh'].min(), df_home['Vitvaror_Energi_Wh'].max())
```
